```python
import pandas as pd
import os
import shutil
```

```python
# Create the data for positive samples

FILE_PATH = "Dataset/Covid Positive/metadata.csv"
IMAGES_PATH = "Dataset/Covid Positive/images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR = "Dataset_created/Covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid Folder created")
```

```python
cnt = 0

for (i,row) in df.iterrows():
    if row['finding'] == "Pneumonia/Viral/COVID-19" and row['view']=="PA":
        # PA is a medical term for X-ray front image
        filename = row['filename']
        image_path = os.path.join(IMAGES_PATH, filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        # print("Moving Image",cnt)
        cnt += 1
        
print(cnt)
```

```python
# Sampling of images from Kaggle
import random
KAGGLE_FILE_PATH = "Dataset/Covid Negative/train/NORMAL"
TARGET_NORMAL_DIR = "Dataset_created/Normal"
```

```python
image_names = os.listdir(KAGGLE_FILE_PATH)
```

```python
random.shuffle(image_names)
```

```python
for i in range(cnt):
    image_name = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH,image_name)
    
    target_path = os.path.join(TARGET_NORMAL_DIR,image_name)
    
    shutil.copy2(image_path,target_path)
    print("Copying Image",i)
print(i)
```
